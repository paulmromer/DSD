## 7. Cryptography

Someone can authenticate by proving that they know a &ldquo;secret key&rdquo; $$s \in \{0, 1, 2, ... q-1\}.$$

With a digital signature system, one can prove knowledge of $s$ without revealing it.

### 7.1 An Addition-Preserving, One-Way Function 

Suppose that the function $f$ 

1. preserves addition: $f(s) + f(t) = f(s+t);$ and  


7. is a one-way function: given a value $P$, the fastest known way to solve for $s$ is the brute method that tries possible values until it finds one that works.

Then $f$ can be used to create a digital signature system. 

<div style="margin-bottom: 400px">
</div>


### 7.2 A Digital Signature System 

For a message $m,$ a secret $s$, a hash function $H$, and an addition-preserving, one-way function $f$, define:

$$ P = f(s) $$

$$r = H(m | s)$$

$$R = f(r)$$ 

$$c = H(R | P | m)$$

$$z = r + c * s$$ 

$$Z = f(z)$$

The value $P$ is said to be the public key implied by $s$. The secret holder publicizes $P.$

The pair $(R, z)$ is said to be the digital signature of the message $m.$ The secret holder also publicizes the signature of the message. 

The verification equation for the message, $$Z = R + c * P,$$ can be checked using public information. 

To see why the verification equation must hold, apply $f$ to the definition of $z$:

$$ z = r + c * s$$

$$Z = f(z) = f(r + c * s) = f(r) + c * f(s) = R + c * P$$

<div style="margin-bottom: 400px">

</div>


### 7.3 No Impersonation 

An impersonator can pick a random $R.$ This determines $c$ and $Z = R + c * P$. To construct a signature that satisfies the verification equation, the impersonator must then solve the equation $f(z) = Z$ for $z$. 

Because $f$ is a one-way function, if will be infeasible for someone who does not know $s$ to find a signature for $m$ that satisfies the verification equation if $q$ is sufficiently large.

### 7.5 Calculating $f(s)$

To calculate f(s), one must be able to solve an equation of the form $$m * a = t \pmod q$$ for arbitrary $m \in \{0, 1, 2, ... , q-1\}.$
If $q$ is prime, $m^{q-1} = 1 \pmod q,$
so multiply both sides of the equation by $m^{q-2}:$

$$ m^{q-2} * m * a = a = m^{q-2} * t \pmod q$$

Is it feasible to calculate $m^{q-2} \pmod q?$

```python
import secrets
m = secrets.randbits(255)
print(f"{m = }")
```
