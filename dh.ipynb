## Generators and Subgroups

Use a capital letter Z in a "blackboard bold typeface" to denote the group of integers mod 22: $$\mathbb{Z}_{22}$$ 

For an element $g \in \mathbb{Z}_{22},$ the subgroup $S$ generated by $g$ can be written as 

$$S = \{ n * g\bmod{22} \text{ : }\  n \in \mathbb{Z} \}.$$

In Python, use this function to calculate the subgroup for any generator: 

```python
def sub_61(g):
    return {(n * g) % 61 for n in range(62)}
```

```python
def sub_63(g):
    return {(n * g) % 63 for n in range(63)}
```

```python
for i in range(63):
    print(sub_63(i))
```

```python
def sub_64(g):
    return {(n * g) % 64 for n in range(64)}
```

```python
def subgroups_61():
    subgroups = set()
    for m in range(61):
        subgroups.add(frozenset(sub_61(m)))
    return subgroups


for subg in subgroups_61():
    print(subg)
```

```python
def subgroups_63():
    subgroups = set()
    for m in range(61):
        subgroups.add(frozenset(sub_63(m)))
    return subgroups


for subg in subgroups_63():
    print(subg)
```

```python
def subgroups_64():
    subgroups = set()
    for m in range(61):
        subgroups.add(frozenset(sub_64(m)))
    return subgroups


for subg in subgroups_64():
    print(subg)
```

## An Addition-Preserving Function

Pick some very large prime number. One we will use frequently is $$p = 2^{255} - 19.$$ Mathematicians have established that this is indeed a prime number. Pick some element in the group $\mathbb{Z}_p$ to be a generator. For example, $g = 7.$ Then we can define a function $$f: \mathbb{Z}_p \rightarrow \mathbb{Z}_p$$ by the rule $$f(z) = z * 9 \ \verb|%|\ p.$$

One of the reasons why modular arithmetic is so interesting is that calculating a residual preserves addition. For any values $y, z \in \mathbb{Z}_p,$ $$ (y + z)\ \verb|%|\ p = ((y \ \verb|%|\  p) + (z \ \verb|%|\  p)) \ \verb|%|\  p .$$

Even better, it also preserves multiplication. 
$$ (y * z)\ \verb|%|\ p = (y \ \verb|%|\  p) * (z \ \verb|%|\  p).$$
It follows that the function $f$ also preserves addition: 

$$ f(y + z)\ \verb|%|\ p = f(y) \ \verb|%|\  p +  f(z) \ \verb|%|\  p.$$

This is half of what we need for a cryptosystem. The other half is for $f$ to be a one-way function. At first, you might think that this will be true. 

Even for a very large numbers $z$ it is easy to calculate $f(z)$. Try it. You'll see. You can use the function randint from the random library to give you a large random integer and you can apply the function $f$.  

```python
from random import randint

p = 2**255 - 19

z = randint(0, p-1)

x = 9 * z % p
x
```

For $f$ to be a one-way function, it must be easy to calculate something like $f(z)$ and it must be impossibly difficult to go the other direction. Given the value for $f(z),$ it must be impossible in practice to calculate the value $z$ even for someone who has access to the most sophisticated computers ever made. 

Don't get your hopes up. A French mathematician discovered something that makes it very easy to go backwards from $f(z)$ to $z.$ In his honor, it is known as Fermat's Little Theorem. It shows that for a prime number $p$, and any value $m \in \mathbb{Z}_{p},$

$$ m^{p-1} = 1 \bmod p.$$ 


Now, we can use the fact that remainders preserve multiplication to see that  

$$ \frac{1}{m} = m^{p-2} \bmod p$$

Starting from the value for $x$ given above, we know that z is given by 

$$z = \frac{1}{9} * x \bmod p$$

By Fermat's little theorem we might be tempted to calculate $\frac{1}{9}$ by evaluating the expression

`9**(p-2) % p` 

If you try to evaluate directly, p is so large that you'll end up waiting a very, very long time. 

But here is the really clever and remarkably simple bit of math that you can use to speed the calculation up. You "work your way up a ladder."

```python
def r(n, h):
    print(f"9 to the power 2**{n} is:")
    print(f"{h * h % p}")
    return 2 * n, h * h
```

```python
r(1, 9)
```

```python
r(2, 81)
```

```python
_j, _v = r(4, 6561)
```

```python
_j, _v = r(_j, _v)
```

```python
_j, _v = (1, 9)
for j in range(9):
    _j, _v = r(_j, _v % p)
```
